```python
import matplotlib.pyplot as plt
```

#freefall motion

```python
# initial condition

x = 50
v = 0
g = 9.8
m = 5
```

```python
def updateV(v, f, m, h):
    vNew = v + ((f/m)*h)
    return vNew
```

```python
def updateX(x, v, h, f, m):
    xNew = x + (v*h) + ((f/m)*h*h*0.5)
    return xNew
```

```python
tList, xList, vList = [], [], []

t = 0
h = 0.001
f = -m*g

for i in range(4000):
    x = updateX(x, v, h, f, m)
    v = updateV(v, f, m, h)
    tList.append(t)
    xList.append(x)
    vList.append(v)
    t += h
print(t)
```

```python
plt.plot(tList, xList)
plt.grid()
```

```python
plt.plot(tList, vList)
plt.grid()
```

```python
plt.plot(tList[::10], xList[::10])
plt.grid()
```

```python
plt.plot(tList[::10], vList[::10])
plt.grid()
```
