- try to simulate a damp oscillation euler method

```python
import matplotlib.pyplot as plt
import math
```

```python
class System:
    def __init__(self):
        #initial method
        self.k = 50
        self.v = 1
        self.m = 10

        self.x = 0.5
        self.f = -(self.k*self.x)
        self.fd = 1
        self.h = 0.001
        self.t = 0

    def updateF(self):
        self.f = -(self.k*self.x)+(-self.fd*self.v*self.v)

    def updateV(self):
        self.v = self.v + ((self.f/self.m)*self.h)

    def updateX(self):
        self.x = self.x + (self.v*self.h) + (0.5*(self.f/self.m)*self.h*self.h)

    def updateT(self):
        self.t += self.h
```

```python
tLs = []
xPos = []

s = System()

for i in range(300000):
    tLs.append(s.t)
    xPos.append(s.x)

    s.updateT()
    s.updateF()
    s.updateV()
    s.updateX()
```

```python
plt.plot(tLs, xPos)
plt.ticklabel_format(style="plain")
```
